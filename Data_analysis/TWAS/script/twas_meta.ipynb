```r
library(tidyverse)
library(data.table)

analysis_region <- fread('~/rl3328/gpQTL_project/finemapping/script/TADB_enhanced_cis.coding_nochrX.bed')
analysis_region %>% head

gene_tss <- fread('~/rl3328/resource/Homo_sapiens.GRCh38.103.chr.reformatted.collapse_only.gene.region_list.2025')
gene_tss %>% head

weights_files <- list.files('/mnt/lustre/data/gaowang/twas_weights/ROSMAP_eQTL', 
                           pattern = 'rds$', 
                           full.names = TRUE, 
                           recursive = TRUE)
weights_files <- grep("DeJager", weights_files, value = TRUE)
weights_files %>% length

data_meta <- data.frame( original_data = weights_files) %>% 
    mutate(gene_id = str_extract(original_data, 'ENSG[0-9]+')) %>%
    merge(analysis_region) %>% 
    merge(gene_tss[,c('TSS', 'gene_id')]) %>% 
    mutate(contexts = NA) %>% 
    rename(region_id = gene_id) %>% 
    select(`#chr`, region_id, TSS, start, end, contexts, original_data) %>% distinct
write.table(data_meta, "./ROSMAP_eQTL.meta_twas.tsv", sep=" ", quote=TRUE, col.names=TRUE, row.names=FALSE) 
```
